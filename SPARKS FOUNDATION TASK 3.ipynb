## DATA VISUALIZATION

```python
# Drawing Heatmap to identify correlation between numerical columns:
plt.figure(figsize=(10,8))
correlation=df.corr()
sns.heatmap(correlation,annot=True,cmap="Blues")
plt.show()
```

Inference:

1) From the heatmap, we can see that there is positive correlation of 0.48 between Sales and Profit. Thus if Sales increases, Profit increases.

2) From the heatmap, we can see that there is negative correlation of -0.22 between Profit and Discount. Thus if Discount increases, most of the times Profit goes down.

```python
# Calculating the statewise profit
profit_sum = df.groupby("State")["Profit"].sum()
profit_sum
```

```python
# State wise Profit of United States of America
plt.figure(figsize=(20,15))
sns.barplot(x=profit_sum.index, y=profit_sum.values)
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

Inference:

1) Highest profit is shown by 'California' and lowest profit is shown by 'Texas'

2) Top 4 profit giving states: 'California','New York','Washington'and'Michigan'

3) Top 4 loss giving states:'Illinois','Pennsylvania','Ohio'and'Texas'
